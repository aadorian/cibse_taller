```
# Bibliotecas a utilizar

import pandas as pd
from ydata_profiling import ProfileReport
```

## Cargamos los archivos de datos csv a Pandas

Los datos se cargan en estructuras llamadas DataFrame para poder trabajar con ellos.

```
# Cargar el archivo CSV de Emisivos en un DataFrame
url_emisivos = 'https://raw.githubusercontent.com/aadorian/cibse_taller/main/datasets/emisivos.csv?token=GHSAT0AAAAAACBPEN3A37TEGHWPG7U2LDHQZCYIGJA'


emisivos = pd.read_csv(url_emisivos, sep=';')
emisivos
```

```
url_operadores= 'https://raw.githubusercontent.com/aadorian/cibse_taller/main/datasets/operadores.csv?token=GHSAT0AAAAAACBPEN3BVO5EW2AUYYIRB4ICZCYIJSQ'
operadores = pd.read_csv(url_operadores, sep=';')
operadores
```

```
# Cargar el archivo CSV de Receptivos en un DataFrame
url_receptivos = 'https://raw.githubusercontent.com/aadorian/cibse_taller/main/datasets/receptivos.csv?token=GHSAT0AAAAAACBPEN3AO2VIJXCOBSBNMKHSZCYIIAA'


receptivos = pd.read_csv(url_receptivos, sep=',')
receptivos
```

## Data Profiling

```
# Profiling básico

print("DESCRIBE:\n")
print(emisivos.describe())
print("\n--------------------------------------------------------------------------------\n")
print("INFO:\n")
print(emisivos.info())

print("DESCRIBE:\n")
print(receptivos.describe())
print("\n--------------------------------------------------------------------------------\n")
print("INFO:\n")
print(receptivos.info())

print("DESCRIBE:\n")
print(operadores.describe())
print("\n--------------------------------------------------------------------------------\n")
print("INFO:\n")
print(operadores.info())
```

```
report_emi = ProfileReport(emisivos,title='Data Profiling Emisivos', minimal=True)
report_emi
report_emi.to_file("emisivos.html")
```

```
report_rec = ProfileReport(receptivos,title='Data Profiling Receptivos', minimal=True)
report_rec
report_rec.to_file('receptivos.html')
```

```
report_operadores = ProfileReport(operadores, minimal=True)
report_operadores
report_operadores = ProfileReport(operadores, title='Data Profiling Operadores', explorative=True)
report_operadores.to_file('operadores.html')
```
